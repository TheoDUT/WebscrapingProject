```python
import pandas as pd
df_weather = pd.read_csv("daily_data.csv")
df_weather
```

```python


# Conversion de la colonne 'date' en datetime sans décalage horaire
df_weather['date'] = pd.to_datetime(df_weather['date']).dt.tz_localize(None)

# Définitions des saisons avec des dates de début et de fin basées uniquement sur le mois et le jour
seasons = {
    'Hiver': (('-12-21', '-12-31'), ('-01-01', '-03-20')),  # Hiver chevauche deux années
    'Printemps': ('-03-21', '-06-20'),
    'Été': ('-06-21', '-09-20'),
    'Automne': ('-09-21', '-12-20')
}

# Fonction pour déterminer la saison
def get_season(date):
    # Extraire le mois et le jour de la date
    month_day = date.strftime("-%m-%d")

    for season, period in seasons.items():
        if season == 'Hiver':  # Gestion spéciale pour l'hiver (2 périodes)
            start1, end1 = period[0]
            start2, end2 = period[1]

            # Vérifier si la date tombe dans l'une des périodes de l'hiver
            if start1 <= month_day <= end1 or start2 <= month_day <= end2:
                return season
        else:  # Gestion des autres saisons
            start, end = period
            if start <= month_day <= end:
                return season

    return 'Inconnu'

# Appliquer la fonction sur la colonne 'date'
df_weather['saison'] = df_weather['date'].apply(get_season)

# Vérification des valeurs uniques dans la colonne 'saison'
print(df_weather['saison'].value_counts())
```

```python
import pandas as pd
from collections import defaultdict

# Listes des DataFrames pour chaque catégorie
art_list = []
bike_list = []
cruise_list = []
mall_list = []
parks_list = []
tour_list = []
museum_list = []

# Dictionnaire pour regrouper les DataFrames par catégorie
category_dfs = defaultdict(list)

# Créer les DataFrames et les ajouter dans les listes appropriées
for index, row in df_establishment.iterrows():
    # Nom du DataFrame basé sur la colonne Name
    nom_df = f"df_{row['Name'].replace(' ', '_')}"
    nom_df = nom_df.replace('\'', '').replace('.', '').replace('-', '_')

    # Créer un DataFrame pour chaque établissement
    establishment_df = pd.DataFrame(row["Reviews"], columns=["Reviewer", "Date", "Rating", "Review_Text"])

    # Ajouter une colonne pour le nom de l'établissement
    establishment_df['Etablissement'] = row['Name']

    # Ajouter le DataFrame à la catégorie appropriée
    if row["Category"] == "art":
        art_list.append(establishment_df)
        category_dfs["art"].append(establishment_df)
    elif row["Category"] == "bike":
        bike_list.append(establishment_df)
        category_dfs["bike"].append(establishment_df)
    elif row["Category"] == "cruise":
        cruise_list.append(establishment_df)
        category_dfs["cruise"].append(establishment_df)
    elif row["Category"] == "mall":
        mall_list.append(establishment_df)
        category_dfs["mall"].append(establishment_df)
    elif row["Category"] == "parks":
        parks_list.append(establishment_df)
        category_dfs["parks"].append(establishment_df)
    elif row["Category"] == "tour":
        tour_list.append(establishment_df)
        category_dfs["tour"].append(establishment_df)
    elif row["Category"] == "museum":
        museum_list.append(establishment_df)
        category_dfs["museum"].append(establishment_df)

# Concaténer tous les DataFrames pour chaque catégorie
df_art = pd.concat(art_list, ignore_index=True)
df_bike = pd.concat(bike_list, ignore_index=True)
df_cruise = pd.concat(cruise_list, ignore_index=True)
df_mall = pd.concat(mall_list, ignore_index=True)
df_parks = pd.concat(parks_list, ignore_index=True)
df_tour = pd.concat(tour_list, ignore_index=True)
df_museum = pd.concat(museum_list, ignore_index=True)

# Créer un DataFrame global avec tous les établissements
df_global = pd.concat([df_art, df_bike, df_cruise, df_mall, df_parks, df_tour, df_museum], ignore_index=True)

df_global

```

```python
# Fonction pour déterminer la valeur de sentiment_5_output_traite
def determine_sentiment_5(value):
    if '1' in value:
        return 1
    elif '2' in value:
        return 2
    elif '3' in value:
        return 3
    elif '4' in value:
        return 4
    else:
        return 5

# Appliquer la fonction sur la colonne sentiment_5_output pour créer sentiment_5_output_traite
data['sentiment_5_output_traite'] = data['Sentiment_5_Output'].apply(determine_sentiment_5) -1
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from imblearn.over_sampling import SMOTE

# Seed pour garantir la reproductibilité
seed = 1111

# Définir les variables catégorielles à encoder
categorical_features = ['Etablissement', 'Saison']

# Séparer les features (X) et la cible (y)
X = data_pred.drop(columns=['sentiment_5_output_traite'])
y = data_pred['sentiment_5_output_traite']

# Configurer OneHotEncoder pour les colonnes catégorielles
column_transformer = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first', sparse_output=False), categorical_features)
    ],
    remainder='passthrough'  # Laisser les colonnes non catégorielles inchangées
)

# Encoder les variables catégorielles
X_encoded = column_transformer.fit_transform(X)

# Diviser en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=seed, stratify=y
)

# Appliquer SMOTE sur l'ensemble d'entraînement
smote = SMOTE(random_state=seed)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Configurer le modèle RandomForestClassifier
rf = RandomForestClassifier(random_state=seed)

# Définir les hyperparamètres à tester dans le grid search
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Effectuer un grid search avec validation croisée
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=3,
    scoring='f1',
    verbose=2,
    n_jobs=-1
)

# Entraîner avec les données SMOTE
grid_search.fit(X_train_resampled, y_train_resampled)

# Meilleurs hyperparamètres
best_rf = grid_search.best_estimator_
print("Meilleurs hyperparamètres :", grid_search.best_params_)

# Prédictions sur l'ensemble d'entraînement
y_train_pred = best_rf.predict(X_train_resampled)

# Prédictions sur l'ensemble de test
y_test_pred = best_rf.predict(X_test)

# Afficher le rapport de classification pour l'ensemble d'entraînement
print("Rapport de classification (Entraînement) :\n")
print(classification_report(y_train_resampled, y_train_pred))

# Afficher le rapport de classification pour l'ensemble de test
print("Rapport de classification (Test) :\n")
print(classification_report(y_test, y_test_pred))
```
